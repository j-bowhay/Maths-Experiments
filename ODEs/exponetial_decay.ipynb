\begin{equation}
y^\prime + y = 0, \quad t \in [0, 3], \quad y(0) = 1.
\end{equation}

```julia
using ApproxFun, Plots

d = 0..3
D = Derivative(d)
L = D + 1

A = [L; ivp()]
y = \(A, [0, 1], tolerance=1e-6)
```

```julia
plot(0:0.1:3, y, label="Numerical")
plot!(0:0.1:3, x -> exp(-x), label="Analytical", linestyle=:dash)
```
